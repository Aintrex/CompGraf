```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import numpy as np
import cv2
import matplotlib.pyplot as plt

print("TensorFlow version:", tf.__version__)
```

```python
IMG_WIDTH = 90
IMG_HEIGHT = 140
BATCH_SIZE = 32
EPOCHS = 7

DATASET_DIR = '/root/.cache/kagglehub/datasets/olafkrastovski/handwritten-digits-0-9/versions/2'
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,  # 80% train, 20% val
    rotation_range=5,      # slight rotation
    width_shift_range=0.05,
    height_shift_range=0.05,
    zoom_range=0.05
)

train_generator = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    color_mode='grayscale',
    class_mode='categorical',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    color_mode='grayscale',
    class_mode='categorical',
    subset='validation'
)
```

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 1)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')  # 10 classes
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_generator,
    epochs=7,
    validation_data=val_generator
)
```

```python
# ==============================================
# 6️⃣ Evaluate & Save
# ==============================================
loss, acc = model.evaluate(val_generator)
print(f"Validation accuracy: {acc:.2f}")

model.save('/kaggle/working/digit_cnn_model.h5')
```

```python
import matplotlib.pyplot as plt

# Extract data
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(acc))

# Plot Accuracy
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.title('Training & Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.title('Training & Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')

plt.show()
```

```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from glob import glob

# Load your trained model
model = tf.keras.models.load_model('/kaggle/working/digit_cnn_model.h5')

# Base directory
base_dir = '/root/.cache/kagglehub/datasets/olafkrastovski/handwritten-digits-0-9/versions/2'

# Выбираем случайную папку (0-9)
random_digit = random.randint(0, 9)  # Случайное число от 0 до 9
digit_dir = os.path.join(base_dir, str(random_digit))

print(f"Selected folder: {digit_dir}")

# Получаем все файлы в этой папке
img_paths = glob(os.path.join(digit_dir, '*.jpg'))

if not img_paths:
    # Если нет jpg, ищем png
    img_paths = glob(os.path.join(digit_dir, '*.png'))

if not img_paths:
    print(f"No images found in folder {random_digit}")
else:
    # Выбираем случайный файл
    random_img_path = random.choice(img_paths)
    print(f"Selected file: {os.path.basename(random_img_path)}")

    # Load & preprocess
    img = cv2.imread(random_img_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        print("Error: Could not load image")
    else:
        # OPTIONAL: crop black borders
        _, thresh = cv2.threshold(img, 240, 255, cv2.THRESH_BINARY_INV)
        x,y,w,h = cv2.boundingRect(thresh)
        cropped = img[y:y+h, x:x+w]

        # Resize
        resized = cv2.resize(cropped, (IMG_WIDTH, IMG_HEIGHT))

        # Normalize & reshape
        resized = resized / 255.0
        resized = resized.reshape(1, IMG_HEIGHT, IMG_WIDTH, 1)

        # Predict
        pred = model.predict(resized)
        digit = np.argmax(pred)
        confidence = np.max(pred)

        print(f"True digit: {random_digit}")
        print(f"Predicted digit: {digit} (confidence: {confidence:.2f})")
        print(f"Correct: {digit == random_digit}")

        # Optional: show
        plt.figure(figsize=(8, 4))

        plt.subplot(1, 2, 1)
        plt.imshow(img, cmap='gray')
        plt.title(f"Original Image\nTrue digit: {random_digit}")

        plt.subplot(1, 2, 2)
        plt.imshow(resized.squeeze(), cmap='gray')
        plt.title(f"Processed Image\nPredicted: {digit} (conf: {confidence:.2f})")

        plt.tight_layout()
        plt.show()
```
